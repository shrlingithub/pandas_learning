## 特征工程

### 任务1
* 对目标值进行log转换
* 去除异常值和合并数据

```python
from scipy.stats import norm
```

```python
# log(1+x)转换
train["SalePrice"] = np.log1p(train["SalePrice"])
```

```python
sns.set_style("white")
sns.set_color_codes(palette='deep')
f, ax = plt.subplots(figsize=(8, 7))
#查看新的直方图分布
sns.distplot(train['SalePrice'] , fit=norm, color="b");

# 获取函数的拟合参数
(mu, sigma) = norm.fit(train['SalePrice'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
ax.xaxis.grid(False)
ax.set(ylabel="Frequency")
ax.set(xlabel="SalePrice")
ax.set(title="SalePrice distribution")
sns.despine(trim=True, left=True)

plt.show()
```

```python
# 去除异常值
train.drop(train[(train['OverallQual']<5) & (train['SalePrice']>200000)].index, inplace=True)
train.drop(train[(train['GrLivArea']>4500) & (train['SalePrice']<300000)].index, inplace=True)
train.reset_index(drop=True, inplace=True)
```

```python
# 分割特征和目标值标签
train_labels = train['SalePrice'].reset_index(drop=True)
train_features = train.drop(['SalePrice'], axis=1)
test_features = test
```

```python
# 合并训练集和测试集数据，以便应用特征转换管道到整个数据集
all_features = pd.concat([train_features, test_features]).reset_index(drop=True)
all_features.shape
```

### 任务2
* 填充缺失值，可以尝试使用不同的方式进行填充

```python
def percent_missing(df):
    data = pd.DataFrame(df)
    df_cols = list(pd.DataFrame(data))
    dict_x = {}
    for i in range(0, len(df_cols)):
        dict_x.update({df_cols[i]: round(data[df_cols[i]].isnull().mean()*100,2)})
    
    return dict_x

missing = percent_missing(all_features)
df_miss = sorted(missing.items(), key=lambda x: x[1], reverse=True)
print('Percent of missing data')
df_miss[0:10]
```

```python
sns.set_style("white")
f, ax = plt.subplots(figsize=(8, 7))
sns.set_color_codes(palette='deep')
missing = round(train.isnull().mean()*100,2)
missing = missing[missing > 0]
missing.sort_values(inplace=True)
missing.plot.bar(color="b")

ax.xaxis.grid(False)
ax.set(ylabel="Percent of missing values")
ax.set(xlabel="Features")
ax.set(title="Percent missing data by feature")
sns.despine(trim=True, left=True)
```

```python
# 一些分类特征存储为数值需要转换为字符型数据
all_features['MSSubClass'] = all_features['MSSubClass'].apply(str)
all_features['YrSold'] = all_features['YrSold'].astype(str)
all_features['MoSold'] = all_features['MoSold'].astype(str)
```

```python
def handle_missing(features):
    # 填充NA值
    features['Functional'] = features['Functional'].fillna('Typ')
    features['Electrical'] = features['Electrical'].fillna("SBrkr")
    features['KitchenQual'] = features['KitchenQual'].fillna("TA")
    features['Exterior1st'] = features['Exterior1st'].fillna(features['Exterior1st'].mode()[0])
    features['Exterior2nd'] = features['Exterior2nd'].fillna(features['Exterior2nd'].mode()[0])
    features['SaleType'] = features['SaleType'].fillna(features['SaleType'].mode()[0])
    features['MSZoning'] = features.groupby('MSSubClass')['MSZoning'].transform(lambda x: x.fillna(x.mode()[0]))
    features["PoolQC"] = features["PoolQC"].fillna("None")
    # 用零填充缺失值
    for col in ('GarageYrBlt', 'GarageArea', 'GarageCars'):
        features[col] = features[col].fillna(0)
    # 用None填充缺失值
    for col in ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']:
        features[col] = features[col].fillna('None')
    for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
        features[col] = features[col].fillna('None')
        
    # 根据中位数填充缺失值
    features['LotFrontage'] = features.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))

    objects = []
    for i in features.columns:
        if features[i].dtype == object:
            objects.append(i)
    features.update(features[objects].fillna('None'))
        
    # 对数值特征进行填充
    numeric_dtypes = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    numeric = []
    for i in features.columns:
        if features[i].dtype in numeric_dtypes:
            numeric.append(i)
    features.update(features[numeric].fillna(0))    
    return features

all_features = handle_missing(all_features)
```

## 模型搭建

### 任务1
* 切割训练集和验证集，为了重复试验结果需要设置随机种子
* 创建线性模型或者基于树的模型，并对比其差异
* 改变模型参数再训练模型进行对比

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, train_labels, random_state=0)
```

```python
ridge = Ridge()
```

```python
ridge.fit(X_train, y_train)
```

```python
print("Training set score: {:.2f}".format(ridge.score(X_train, y_train)))
print("Validation set score: {:.2f}".format(ridge.score(X_valid, y_valid)))
```

```python
ridge2 = Ridge(alpha=0.01)
```

```python
ridge2.fit(X_train, y_train)
```

```python
print("Training set score: {:.2f}".format(ridge2.score(X_train, y_train)))
print("Validation set score: {:.2f}".format(ridge2.score(X_valid, y_valid)))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
```

```python
print("Training set score: {:.2f}".format(rfr.score(X_train, y_train)))
print("Validation set score: {:.2f}".format(rfr.score(X_valid, y_valid)))
```

```python
rfr2 = RandomForestRegressor(max_depth=30)
rfr2.fit(X_train, y_train)
```

```python
print("Training set score: {:.2f}".format(rfr2.score(X_train, y_train)))
print("Validation set score: {:.2f}".format(rfr2.score(X_valid, y_valid)))
```
