# ABTest

## ABTest概述

### 什么是ABTest
    ABTest，简单来说，就是为同一个产品目标制定两个方案（比如两个页面一个用红色的按钮、另一个用蓝色的按钮），让一部分用户使用A方案，另一部分用户使用B方案，然后通过日志记录用户的使用情况，并通过结构化的日志数据分析相关指标，如点击率、转化率等，从而得出那个方案更符合预期设计目标，并最终将全部流量切换至符合目标的方案。

### 为什么要进行ABTest
    判断产品的设计：A好还是B好
    判断产品的新功能对KPI影响多大

### 什么时候不适合进行ABTest
    没有清晰的对照组
    很难获得用户的数据

### ABTest的一般步骤
    1.确定测试目标及假设
    2.确定数据指标
    3.确定最小提升值（提升了多少才算有效）
    4.计算用户数量及测试时长
      ● 减少过度资源浪费
      ● 使测试结果更可信

### ABTest中的基本统计学知识
#### 假设检验
    我们观察到的差异比偶然差异大得多，我们就可以认为两组之间的差异是真实存在的
#### 第一类错误、第二类错误
    第一类错误：假阳性；原假设H0符合实际情况，检验结果将它否定了，称为弃真错误。
    第二类错误：假阴性；原假设H0不符合实际情况，检验结果无法否定它，称为取伪错误。


#### 置信水平
    可以接受第一类错误发生的概率（一般α = 0.05）
#### p值
    实际情况中第一类错误发生的概率
#### 统计功效
    1-第二类错误发生的概率（0.8）
    （为什么叫功效？）

### ABTest注意事项
    数据可靠性检查
        - 检查一些不该因为AB测试而产生变化的指标，并弄清原因
    显著性测试
        - 如果AB测试结果显著而全局实施新的版本结果不显著，需要对用户进行更一步细分，检查测试的设计是否有问题
    不要过度相信定义的指标
        - 指标未必能够反映实际情况
        - 指标之间可能相互抵触
    避免辛普森悖论
        - 在AB测试的结果分析上进行多维度细分对比
        （两个表现得好的指标，合起来不一定就好）

```python
# 背景
# 目前的conversion是13%，希望在新产品上能提升2%，达到15%
# 控制组：旧产品
# 测试组：新产品
import numpy as np
import pandas as pd 
import scipy.stats as stats
import statsmodels.stats.api as sms
import matplotlib as mpl
import seaborn as sns
from math import ceil

# 选择样本大小： 样本过多->浪费资源；样本太少->不足以满足测试要求
# power analysis 能效分析
effect_size = sms.proportion_effectsize(0.13,0.15)
# 计算样本量
required_n = sms.NormalIndPower().solve_power(
    effect_size,
    power=0.8,
    alpha=0.05,
    ratio=1
)
# 取整
required_n = ceil(required_n)
print(required_n)
```
